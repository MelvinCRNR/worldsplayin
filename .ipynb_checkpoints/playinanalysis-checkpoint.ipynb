```python
df=pd.read_csv("2021playin.csv");
df.info();
pd.set_option("max_rows", None);
"""df.sort_values(by = ['Team', 'Opponent'], ascending = True)"""
df
```

```python
gold_team=df.loc[:,['Team','Gold Earned']];
gold_team=gold_team.sort_values(by = 'Team', ascending = True);
gold_team_mean=gold_team.groupby('Team').mean();
gold_team_mean.columns = ['Mean Gold/Player/Team'];
gold_team_mean=gold_team_mean.sort_values(by = 'Mean Gold/Player/Team', ascending = True);
gold_team_mean
```

```python
sns.catplot(data=df, x='Team', y='Gold Earned', kind='bar', height=8.27, aspect=11.7/8.27).set_xticklabels(rotation=90);
```
