```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
print("loaded")
```

```python
df=pd.read_csv("2021playin.csv");
df.info();
pd.set_option("max_rows", None);
"""df.sort_values(by = ['Team', 'Opponent'], ascending = True)"""
df
```

```python
df.drop(df.index[200:220],0,inplace=True);
df.info();
df
```

```python
df['Team'].value_counts()
```

```python
gold_team=df.loc[:,['Team','Gold Earned']];
gold_team=gold_team.sort_values(by = 'Team', ascending = True);
gold_team_mean=gold_team.groupby('Team').mean();
gold_team_mean.columns = ['Mean Gold/Player'];
gold_team_mean=gold_team_mean.sort_values(by = 'Mean Gold/Player', ascending = True);
gold_team_mean
```

```python
gold_damage=df.loc[:,['Gold Earned','Champion Damage Share']];
gold_damage
```
